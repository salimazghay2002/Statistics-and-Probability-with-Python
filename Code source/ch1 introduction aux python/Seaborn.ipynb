```python
import seaborn as sns
data = sns.load_dataset("tips")
```

```python
# #Affichage des 5 premi√®res lignes de les donnees
data.head()
```

```python
#afficher un histogramme de la distribution des donn√©es dans la colone : " smoker "
sns.histplot(data['smoker'])
```

```python
#cr√©ation un nuage de points pour visualiser la relation entre " sexe " et " facture " (total_bill).
sns.scatterplot(x='sex',y='total_bill',data=data)
```

```python
#cette commande affiche un diagramme en boite pour visualiser la distribution des valeurs num√©rique en fonction des cat√©gories
sns.boxplot(x='tip',y='day',data=data)
```

```python
# cr√©ation d'une carte de chaleur(heatmap)pour visualiser la corr√©lation entre les variables
corr=data.corr()
sns.heatmap(corr, annot=True)
```
