```python
import seaborn as sns
data = sns.load_dataset("tips")
```

```python
# #Affichage des 5 premières lignes de les donnees
data.head()
```

```python
#afficher un histogramme de la distribution des données dans la colone : " smoker "
sns.histplot(data['smoker'])
```

```python
#création un nuage de points pour visualiser la relation entre " sexe " et " facture " (total_bill).
sns.scatterplot(x='sex',y='total_bill',data=data)
```

```python
#cette commande affiche un diagramme en boite pour visualiser la distribution des valeurs numérique en fonction des catégories
sns.boxplot(x='tip',y='day',data=data)
```

```python
# création d'une carte de chaleur(heatmap)pour visualiser la corrélation entre les variables
corr=data.corr()
sns.heatmap(corr, annot=True)
```
