```python
import matplotlib.pyplot as plt
import numpy as np

# Génération de données aléatoires
np.random.seed(123)
a = np.random.normal(0, 1, size=500)

# Création de l'histogramme
plt.hist(a, bins=20, density=True, alpha=0.5)

# Ajout des étiquettes des axes et du titre
plt.xlabel('Valeurs')
plt.ylabel('Densité de probabilité')
plt.title('Histogramme de valeurs aléatoires')

# Affichage de la figure
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def DiagrammeEnSecteurs(x,y):
    # Creating plot
    fig = plt.figure(figsize=(10, 7))

    def func(pct, allvalues):
        absolute = int(pct / 100. * np.sum(allvalues))
        return "{:.1f}%\n".format(pct, absolute)

    plt.pie(y, autopct=lambda pct: func(pct, y),labels=x)
    # show plot
    plt.show()
x = ['Fraises', 'Bananes', 'Pommes', 'Oranges']
y = [15, 35, 40, 10] # Pourcentage de chaque catégorie
DiagrammeEnSecteurs(x,y)
```
