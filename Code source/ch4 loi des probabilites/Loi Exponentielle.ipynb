```python
# Manipulation de loi exponentielle
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon
```

```python
# Paramètre de la loi exponentielle
lambda_param = 0.5 # Paramètre lambda (taux)

# Génération de données aléatoires selon la loi exponentielle
data = expon.rvs(scale=1/lambda_param, size=1000)

# Calcul de la probabilité d'attendre un certain temps avant qu'un événement se produise
temps_attente = 2
prob = expon.pdf(temps_attente, scale=1/lambda_param)
print("La probabilité d'attendre {} unités de temps est de {:.2f}".format(temps_attente, prob))

# Calcul de l'espérance et de la variance
esperance = expon.mean(scale=1/lambda_param)
variance = expon.var(scale=1/lambda_param)
print("L'espérance de la loi exponentielle est E(X) = {} et la variance est V(X) = {}".format(esperance, variance))
```

```python
# Tracé de la distribution de probabilité
x = np.linspace(0, 10, 100)
y = expon.pdf(x, scale=1/lambda_param)
plt.plot(x, y, label='Distribution de probabilité')

# Tracé d'un autre graphique sur la même courbe
y2 = expon.pdf(x, scale=0.2)
plt.plot(x, y2, label='Autre courbe')
plt.title("Distribution de probabilité de la loi exponentielle")
plt.xlabel("Temps")
plt.ylabel("Probabilité")
plt.legend()
plt.show()
```
