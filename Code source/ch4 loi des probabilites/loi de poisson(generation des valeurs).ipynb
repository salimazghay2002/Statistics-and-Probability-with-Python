```python
import matplotlib.pyplot as plt

# Paramètre de la loi de Poisson
lam = 3

# Génération de 1000 valeurs aléatoires suivant une loi de Poisson
x = np.random.poisson(lam, size=1000)

# Affichage des valeurs sous forme d'un histogramme
plt.hist(x, bins=20)
plt.xlabel('Valeurs')
plt.ylabel('Fréquence')
plt.title('Histogramme des valeurs générées aléatoirement suivant une loi de Poisson')
plt.show()
```
