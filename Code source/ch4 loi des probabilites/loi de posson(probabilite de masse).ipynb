```python
from scipy.stats import poisson

lambda_ = 1  # paramètre lambda
k = 3         # valeur pour laquelle on veut calculer la probabilité de masse

# Calcul de la probabilité de masse
pmf = poisson.pmf(k, lambda_)
pmf
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Paramètres de la loi de Poisson
lambda_ = 2

# Domaine de définition de la probabilité de masse
k = np.arange(0, 20)

# Calcul de la probabilité de masse pour chaque valeur de k
pmf = poisson.pmf(k, lambda_)

# Tracé du graphique de la probabilité de masse
plt.plot(k, pmf)

# Ajout de légendes et de titres
plt.xlabel('k')
plt.ylabel('P(X=k)')
plt.title('Probabilité de masse de la loi de Poisson')

# Affichage du graphique
plt.show()
```
