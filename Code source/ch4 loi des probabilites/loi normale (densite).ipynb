```python
import scipy.stats as stats

# Définir les paramètres de la loi normale
mu = 1
sigma = 2

# Calculer la densité de probabilité pour x = 0
x = 0
pdf = stats.norm.pdf(x, mu, sigma)

# Afficher la densité de probabilité
print(f"La densité de probabilité pour x = {x} est {pdf:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-6, 8, 100)
y=stats.norm.pdf(x,1,2)
plt.plot(x,y)
plt.ylim(0,0.2)
plt.title('densite de la loi normale N(1,2)')
plt.xlabel("x")
plt.grid()
plt.xlim(-6,8)

# Afficher le graphique
plt.show()
```

```python
from scipy.stats import norm

# Calculer les valeurs de densité de probabilité pour x = 1 et 2.
x1 = norm.pdf(1)
x2 = norm.pdf(2)


# Afficher les résultats
print(f"La valeur de densité de probabilité pour x = 1 est : {x1}")
print(f"La valeur de densité de probabilité pour x = 2 est : {x2}")
```

```python
# Définir l'intervalle de x
x = np.linspace(-6, 6, 100)

# Calcul la densité de probabilité pour chaque valeur de x
y = norm.pdf(x)

# Afficher la courbe de densité de probabilité
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('Densité de probabilité')
plt.title('Loi normale centrée réduite')
plt.show()
```
