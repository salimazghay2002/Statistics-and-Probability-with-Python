```python
from scipy.stats import norm
p=norm.cdf(0, 2, 3)
p
```

```python
import numpy as np
x = np.array([-2, -1, 0, 1, 2])
p = norm.cdf(x, 0, 1)
p
```

```python
import matplotlib.pyplot as plt

mu, sigma = 1, 2  # paramètres de la loi normale
x = np.linspace(-5, 5, 100)
y = norm.cdf(x, mu, sigma)

plt.plot(x, y)
plt.title(r"Fonction de répartition de la loi normale : ")
plt.xlabel("X")
plt.ylabel("F(X)")
plt.xlim(-5,5)
plt.show()
```

```python
# Définir une plage de valeurs de x
x = np.linspace(-6, 7, num=100)

# Calculer les valeurs de la fonction de répartition pour chaque valeur de x
phi_x = norm.cdf(x)

# Tracer le graphique de la fonction de répartition
plt.plot(x, phi_x)
plt.xlabel('x')
plt.xlim(-6,7)
plt.ylabel('$\Phi(x)$')
plt.title('Fonction de répartition de la loi normale centrée réduite')
plt.show()
```
