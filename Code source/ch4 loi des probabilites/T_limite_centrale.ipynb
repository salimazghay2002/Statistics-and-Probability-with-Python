```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def histogramme(n,m,mu):
    sample_means = []  # Liste pour stocker les moyennes des échantillons

    for _ in range(m):
        sample = np.random.exponential(mu, n)  # Génération d'un échantillon de taille n
        sample_mean = np.mean(sample)  # Calcul de la moyenne de l'échantillon
        sample_means.append(sample_mean)

    # Tracé de l'histogramme des moyennes des échantillons
    plt.hist(sample_means, bins=30, density=True, alpha=0.5)
    plt.xlabel('Valeurs')
    plt.ylabel('Fréquence')
    plt.show()
```

```python
# Génération d'un grand nombre de variables aléatoires exponentielles
n = 1000  # Taille de l'échantillon
m = 10000  # Nombre d'échantillons
mu = 2  # Moyenne de la distribution exponentielle

histogramme(n,m,mu)
```
