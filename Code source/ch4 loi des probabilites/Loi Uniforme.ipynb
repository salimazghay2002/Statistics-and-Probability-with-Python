```python
# Manipulation et simulation de loi Uniforme
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform
```

```python
# Paramètres de la loi uniforme
a = 2 # Limite inférieure
b = 5 # Limite supérieure

# Calcul de la probabilité pour X entre 3 et 8
prob = uniform.cdf(8, loc=a, scale=b-a) - uniform.cdf(3, loc=a, scale=b-a)
print("La probabilité que X soit entre 3 et 8 est de {:.2f}%".format(prob * 100))

# Calcul de l'espérance et de la variance
esperance = uniform.mean(loc=a, scale=b-a)
variance = uniform.var(loc=a, scale=b-a)
print("L'espérance de X est E(X) = {:.2f} ".format(esperance))
print("La variance est V(X) = {:.2f}".format(variance))
```

```python
# Tracé de la densité de probabilité et de la fonction de répartition
x = np.linspace(a, b, 100)
y_pdf = uniform.pdf(x, loc=a, scale=b-a)
y_cdf = uniform.cdf(x, loc=a, scale=b-a)
plt.plot(x, y_pdf, label='Densité de probabilité')
plt.plot(x, y_cdf, label='Fonction de répartition')
plt.title("Loi uniforme - Densité de probabilité et Fonction de répartition")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
