```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Définir les donnez de population
data = pd.read_csv("adult.csv")
print(data['hours.per.week'])
```

```python
 # Utiliser la méthode cut() pour créer la distribution
# statistique groupée
ventes_groupées = pd.cut(data['hours.per.week'], bins=10, right=False)
# Afficher la distribution statistique groupée
print(ventes_groupées.value_counts())
```

```python
# Tracer l'histogramme
plt.hist(data['hours.per.week'], bins=10, alpha=0.5, color='steelblue',
edgecolor='black')

# Ajouter des titres et des étiquettes
plt.title('histogramme')
plt.xlabel('hours.per.week')
plt.ylabel('Fréquence')

# Afficher le graphique
plt.show()
```
