```python
# importer les bibliotique necessaires
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
# Dictionnaire de donn√©es
donnees = {
'M2': [100, 105, 85, 90, 125, 123, 130 ,105 , 130,90 , 91, 95, 200],
'annees': [2010,2011, 2013, 2015, 2016, 2017 , 2018, 2014, 2019, 2020, 2021, 2022, 2023],
'prix': [200000, 210000, 210000, 230000, 270000,300000, 310000, 320000, 330000, 330000, 350000, 360000, 400000]}
```

```python
# repr√©sentation graphique de corr√©lation en utilisant seaborn
data = pd.DataFrame(donnees)
sns.pairplot(data)
```

```python
# la corr√©lation entre les variables
corr=data.corr()
sns.heatmap(corr, annot=True)
```

```python
# Convertir les donn√©es en tableaux NumPy
X = np.array([donnees['M2'], donnees['annees']]).T
y = np.array(donnees['prix'])
```

```python
# Cr√©er le mod√®le de r√©gression lin√©aire multiple
regression = LinearRegression()
# Entra√Æner le mod√®le
regression.fit(X, y)
```

```python
# Coefficients de r√©gression
coefficients = regression.coef_
intercept = regression.intercept_
```

```python
# Affichage des coefficients et de l'intercept
print("Coefficients:", coefficients)
print("Intercept :", intercept)
```

```python
#Pr√©diction
nouvelles_donnees = {'M2': [110, 130, 95], 'annees': [2024, 2025, 2026] }
X_nouveau = np.array([nouvelles_donnees['M2'],nouvelles_donnees['annees']]).T
predictions = regression.predict(X_nouveau)
print("Pr√©dictions :", predictions)
```
